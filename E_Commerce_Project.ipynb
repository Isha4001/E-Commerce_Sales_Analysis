```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = 'plotly_white'
```

```python
data = pd.read_csv("Superstore_sample.csv", encoding = 'latin-1')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

# Converting Date column

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# Monthly Sales Analysis

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month,x='Order Month',y='Sales',title='Montly Sales Analysis')
fig.write_html("monthly_sales.html")
```

# Sales by Category

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
fig = px.pie(sales_by_category,values='Sales',names='Category',hole=0.5,title='Sales by Category',color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside',textinfo='percent+label')
fig.update_layout(title_font=dict(size=24))
fig.show()
```

# Sales by Sub-Category

```python
sales_by_subcat = data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_subcat,x='Sub-Category',y='Sales',title='Sales by Sub Category')
fig.show()
```

# Profit Analysis by Month

```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
fig = px.line(profit_by_month,x='Order Month',y='Profit',title='Profit by Month')
fig.show()
```

# Profit Analysis by Category

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
fig = px.pie(profit_by_category,values='Profit',names='Category',hole=0.5,color_discrete_sequence=px.colors.qualitative.Pastel,title='Profit by Category')
fig.update_traces(textposition = 'outside',textinfo = 'percent+label')
fig.update_layout(title_font=dict(size=24))
fig.show()
```

# Profit Analysis by Sub Category

```python
profit_by_subcat = data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_subcat,x='Sub-Category',y='Profit',title='Profit by Sub Category')
fig.show()
```

# Sales & Profit - Customer Segment

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                    marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment', yaxis_title='Amount')
fig.show()
```

# Sales to Profit Ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio']=sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales_to_Profit_Ratio']])
```
